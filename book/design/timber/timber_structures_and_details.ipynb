```{admonition} Rules of thumb
**Solid timber**  
$l = 5 - 25 m$  
$s = 2.5 - 6 m \rightarrow  h = \frac{l}{10} - \frac{l}{14}$  

**Glulam**  
$l = 20 - 80 m$  
$s = 2.5 - 6 m \rightarrow  h = \frac{l}{10} - \frac{l}{15}$
```

```{figure} Images/straight_truss_shapes.png
---
---
Possible shapes.
```

```{figure} Images/straight_truss_details.png
---
---
Details (left: connection of tensile bars (steel strips) to bottom chord and compressive strut (coupler plate); middle: connection between bottom chord and diagonals with screwed steel plates and central axle; right: connection between bottom chord and diagonals with screwed steel plates and central axle).
```

The upper chords (under compression) must be supported laterally in all trusses. For truss type 1 the bottom chord also needs lateral support to prevent the tensile chord from twisting out of plane.

## Triangular or trapezoidal truss

```{figure} Images/triangular_truss.png
---
---
Triangular or trapezoidal truss.
```

## Three hinged frame (curved/kinked legs)

```{figure} Images/three_hinged_kinked.png
---
---
Three hinged frame (curved/kinked legs).
```

```{admonition} Rules of thumb
**Frame with kinked corners**  
$l = 15 - 50 m$  
$\alpha = 20 ^{\circ}$  
$s = 4 - 5 m \rightarrow h_{kink} = \frac{l + g}{28} and h_{foot} = 0.4 \cdot h_{kink}$  
$s = 5 - 9 m \rightarrow h_{kink} = \frac{l + g}{24} and h_{foot} = 0.4 \cdot h_{kink}$  

**Frame with bent corners**  
$l = 15 - 50 m$  
$r = 3.5 - 5 m$  
$\alpha = 20 ^{\circ}$
$s = 4 - 5 m \rightarrow h_{kink} = \frac{l + g}{31} and h_{ridge} = 0.5 \cdot h_{kink} and h_{foot} = 0.67 \cdot h_{kink}$  
$s = 5 - 9 m \rightarrow h_{kink} = \frac{l + g}{26} and h_{ridge} = 0.5 \cdot h_{kink} and h_{foot} = 0.67 \cdot h_{kink}$  
```

## Arch (solid and truss beam)

```{figure} Images/arch.png
---
---
Arch (solid and truss beam).
```

```{admonition} Rules of thumb
**Glulam beam**  
$l = 20 - 100 m$  
$f = \frac{l}{5} - \frac{l}{7}$ 
$s = \leq 5 m \rightarrow h = \frac{l}{45}$  
$s = 5 - 9 m \rightarrow h = \frac{l}{40}$  

**Glulam truss**  
$l = 50 - 120 m$  
$f = \frac{l}{5} - \frac{l}{8}$ 
$s = 5 - 10 m$  
$h = \frac{l}{20} - \frac{l}{40}$
```
