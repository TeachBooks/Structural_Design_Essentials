$$
M = factor * F * L
$$  

$$
R = factor * F
$$

case numbers in the tables are related to the concentrated loads working on the structure. The bending moments and support reactions in the designated points are found by multiplying the factors in the table with F x L or F.

- M(1) is the maximum bending in beam (1).
- M3 is the hogging moment located at support 3.
- R2 is the support reaction in support 2.


`````{admonition} Example
:class: tip
For situation c:
- $F = 30 kN$
- $L = 8 m$

The calculations are as follows:
- $M(1) = 0.175 * 30 * 8 kNm$
- $M3 = -0.15 * 30 * 8 kNm$
- $R2 = 1.15 * 30 kN$
`````

## Beams on multiple supports (distributed load)


```{figure} Images/Beams_On_Multiple_Supports2.png
---
scale: 100%
---
Beams on multiple supports for distributed loads.
```

```{table} Bending moments and support reactions for beams on multiple supports with distributed load.
:name: bending-and-reaction-force-beam-multiple-supports-distributed-load
| Case | M(1) | location rel. to 1* | M2     | M(2) | location rel. to 2** | M3     | M(3) | R1     | R2     | R3     | R4     |
|------|----------------|-----------|--------|-----------------|-----------|--------|------|--------|--------|--------|--------|
| h    | 0.0703                    | 0.3750 | -0.1250                    | 0.0703 | 0.6250||| 0.3750 | 1.2500 | 0.3750 |        |
| i    | 0.0957                    | 0.4375 | -0.0625                    | -0.0313| ||      | 0.4375 | 0.6250 | -0.0625|        |
| j    | 0.0800                    | 0.4000 | -0.1000                    | 0.0250 | 0.5000| -0.1000| 0.0800 | 0.4000 | 1.1000 | 1.1000 | 0.4000 |
| k    | 0.1013                    | 0.4500 | -0.0500                    | -0.0500|       | -0.0500| 0.1013 | 0.4500 | 0.5500 | 0.5500 | 0.4500 |
| l    | -0.025                    |        | -0.0500                    | 0.0750 | 0.5000| -0.0500|        | -0.0500| 0.5500 | 0.5500 | -0.0500|
| m    | 0.0735                    | 0.3833 | -0.1167                    | 0.0535 | 0.5830| -0.0333|        | 0.3833 | 1.2001 | 0.4499 | -0.0333|
| n    | 0.0939                    | 0.4333 | -0.0667                    | -0.0250|       | 0.0167 |        | 0.4333 | 0.6501 | -0.1001| 0.0167 |
```

$$
M = factor * q * L^2
$$

$$
R = factor * q * L
$$

*) location of bending moment M(1) in relation to support 1 (distance = factor x L)  
**) location of bending moment M(2) in relation to support 2 (distance = factor x L) 

The numbers in the tables are related to the concentrated loads
working on the structure. The bending moments and support reactions in
the designated points are found by multiplying the factors in the table with
$q * L^2$ or $q * L$.

- M(1) is the maximum bending in beam (1).  
- M3 is the hogging moment located at support 3.   
- R2 is the support reaction in support 2.   

`````{admonition} Example
:class: tip
For situation m:  
- $q = 5 kN/m$  
- $L = 6 m$  

The calculations are as follows:
- $M(1) = 0,0735 * 5 * 62 \ kNm$  
- $M3 = -0,0333 * 5 * 62 \ kNm$  
- $R2 = 1,2001 * 5 * 6 \ kN$
`````
