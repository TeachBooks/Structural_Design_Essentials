The reliability index $\beta$ can be calculated from the probaility density functions (PDF) of loads (Effects) and resistance, shown in the figure below:

```{Admonition} In depth review of partial factors from Eurocode
:class: note, dropdown

How are the values of $\gamma_{F}$ and $\gamma_{M}$ related to the stochastic spread? From Schneider and Vrouwenvelder [^1] the following equations can be derived:

$$
\gamma_i = \frac{\mu_i - \alpha_i \cdot \beta_0 \cdot \sigma_i}{X_{char,i}}
$$

in which the following variables are visible:
- $\gamma_i$: partial safety factor for i (i can be E or R)
- $\mu_i$: mean value of i
- $\alpha_i$: weighing factor depending on how dominant the factor is for the reliability of the structure, often $\alpha_R = 0.8$ for the most dominant resistance variable, $\alpha_R = 0.32$ for other (non-dominant) resistance variables, $\alpha_E = -0.7$ for the most dominant load variable or $\alpha_E = -0.28$ for non-dominant load variables.
- $\beta_0$: required (targeted) reliability index
- $\sigma_i$: standard deviation of i
- $X_{char,i}$: characteristic value of i

After substitution of E and R, this yields:

$$
\gamma_E = \frac{\mu_E - \alpha_E \cdot \beta_0 \cdot \sigma_E}{E_{char}}
$$

$$
\gamma_M = \frac{R_{char}}{\mu_R - \alpha_R \cdot \beta_0 \cdot \sigma_R}
$$
```

## Effect vs Resistance
The general objective of calculations overseeing safety is that the resistance of the structure should be greater than the effect of the load. A useful way to express this is the *unity check* {eq}`unity_check`. This check is obtained in 4 steps. The image below gives good visual insight in the steps.

$$
\frac{E_d}{R_d} \leq 1
$$(unity_check)

- First, both effect $E$ and resistance $R$ are expressed in their characteristic values. For $E$, this means the value $E_{char}$ is calculated with a probability of exceedance (overshoot) of 5%. For $R$, this is the value with a probability of 5% of under shoot, $R_{char}$. This is a standard statistic operation that is done by adding or subtracting $1.64·σ$ in case of a normal distribution. For loads or materials other distributions may apply, such as Weibull or lognormal distributions.
- Second, on the Effect side, $γ_E$ , the partial load factor, is multiplied with the characteristic value of the load or Effect, resulting in a dimensioning value of $E$ {eq}`E_d`, represented by the blue draw line, and for this example the value $E_d$ = 163 (again, unit does not matter here, as long as both $R$ and $E$ are expressed in the same unit).

$$
E_d = E_{char} \gamma_E
$$(E_d)

- Third, on the Resistance side, the characteristic value of the strength or Resistance is divided by $γ_R$, the partial material factor, resulting in a dimensioning value of $R$ {eq}`R_d`, represented by the red draw line, and for this example the value $R_d = 201$ (same unit as $E$).

$$
R_d = \frac{R_{char}}{\gamma_R}
$$(R_d)

- Fourth and finally, $E_d$ and $R_d$ are compared, in either of the two following ways: direct comparison $E_d ≤ R_d$ or unity check {eq}`unity_check`, the latter with the advantage that it gives some numerical information on the "safety distance" between $E_d$ and $R_d$.


>```{figure} Images/E_R_normaldistributions.png
>---
>---
>Example
>```
>
>Example of a normal distribution of Effects ($E$) and Resistance ($R$) and the place of characteristic values $E_{char}$ and $R_{char}$ and design values $E_d$ and $R_d$. The characteristic values are obtained by finding from the normal distribution the values with a 5% probability of exceedance ($μ_E$ + 1.64 · $σ_E$ for $E$ and $μ_R$ − 1.64 · $σ_R$ for $R$). The design values are found by multiplying /dividing the characteristic values with the partial safety factors $γ_E$ and $γ_R$. The distance on the horizontal axis between $E_d$ and $R_d$ represents a measure of the unity-check (u.c.). In this example, $E_d$ = 0.81 · $R_d$. The structure is more than safe.

## Python implementation
If you want to see some numerical examples using the concepts of safety, you can find these under {ref} `unity_check` and {ref}`reliability_index_example`.




[^1]: Schneider and Vrouwenvelder (2017, equation 4.44)
